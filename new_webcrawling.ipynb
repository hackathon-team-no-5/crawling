```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_0_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

def getPageString(url):
    data = requests.get(url, headers = headers) 
    return data.content

url = 'https://www.koibito.co.kr/product_list_item.html?brand=&item1=%B0%A1%B9%E6&item2=&soldout=&soldout2=&search_sort=price|desc'
```

```python
data = getPageString(url)
parser = BeautifulSoup(data, "html.parser")
find = parser.find("div", {"id":"product_list"})
list = find.findAll("li")
item = list[0]
item
```

```python
name = item.find("div", {"class":"goods_name line_1"})
price = item.find("div", {"class":"goods_price"})
image = item.find("img").get('data-src')
aname = name.text.strip()
aprice = price.getText().strip()
aa=aprice[:10]+'원'
print(aname,aa,image)
```

```python
name_list = []
price_list = []
url_list = []
index_list=[]
for prod in list:
    name = prod.find("div", {"class":"goods_name line_1"}).text.strip()
    pric = prod.find("div", {"class":"goods_price"}).getText().strip()
    price = pric[:10]+'원'
    url = prod.find("img").get('data-src')
    print("이름: {}, 가격: {}, 이미지: {}".format(name,price,url))
    name_list.append(name)
    price_list.append(price)
    url_list.append(url)
```

```python
df = pd.DataFrame({"TITLE" : name_list, "PRICE" : price_list, "URL" : url_list})
df
```
